```python
 # transformation
 index = pd.date_range('10/1/1999', periods=1100)
 ts = pd.Series(np.random.normal(0.5, 2, 1100), index)
 ts.head()
```

```python
ts = ts.rolling(window=100, min_periods=100).mean().dropna()
ts.head()
```

```python
ts.tail()
```

```python
key = lambda x: x.year
```

```python
# We would expect the result to now have mean 0 and standard deviation 1 
# within each group, which we can easily check:
zscored = lambda x: (x - x.mean()) / x.std()
```

```python
ts_zscored = ts.groupby(key).transform(zscored)
ts_zscored.head()
```

```python
ts_zscored.mean()
```

```python
ts_zscored.std()
```

```python
compare = pd.DataFrame({'Original': ts, 'ZScored': ts_zscored})
compare.plot()
```
